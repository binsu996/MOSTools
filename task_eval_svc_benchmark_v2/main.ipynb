```python
import os
from pathlib import Path

def upload_to_s3(f,dry_run=False):
    cmd=f"sudo aws s3 cp '{f}' 's3://wondera-publicmaterials/{f}'"
    url=f'https://wondera-publicmaterials.s3.us-east-2.amazonaws.com/{f}'
    if not dry_run:
        os.system(cmd)
    return url

folders=["prompt","gen_074e38a7","gen_6be84248","gen_a921aa86"]
r={}
for g in folders:
    paths=Path(g).glob("*")
    paths=sorted(paths)
    urls=list(map(upload_to_s3,paths))
    r[g]=urls
    # print(urls)
    # exp.append(urls)
    print(len(urls))
import pandas as pd
df=pd.DataFrame(r)
df
df.to_excel("20240902.xlsx")
```

```python
import numpy as np
import torch
data = np.array([1, 2, 2, 3, 4, 5, 90, 100, 101])
data=torch.Tensor(data)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# 定义离群点的阈值
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 过滤离群点
filtered_data = data[(data >= lower_bound) & (data <= upper_bound)]

# 计算过滤后的最大值、最小值和均值
max_value = np.max(filtered_data)
min_value = np.min(filtered_data)
mean_value = np.mean(filtered_data)
```
